```python
files = glob.glob('./*.bf')
filehash = {}
for f in files:
    temp = pd.read_table(f, index_col=0)
    sample = f.replace('./','')
    sample = sample.replace('.bf','')
    filehash[sample] = temp
frames = []
columns = []
for sample in sorted(filehash.keys()):
    frames.append(filehash[sample]['BF'])
    columns.append(sample)
bf = pd.concat(frames,axis = 1)
bf.columns = columns
bf = bf.drop(['20180111_293a_WT18', '20180111_293a_WT21'], axis=1)
bf = bf[[ 'V2_20170419_T6ABC_WT',
         'V2_20170419_T13ABC_WT',
         'V2_20170419_T16ABC_WT',
         'V2_20170419_T9ABC_FA',
         'V2_20170419_T13ABC_FA',
         'V2_20170419_T17ABC_FA',
         'V2_20170419_T20ABC_FA',
         'V2_20170515_T20ABC_FA',
         'V2_20171215_FA_T17ABC_FA',
         'V2_20171215_FA_T20ABC_FA',
         'V2_20180111_293a_FA19',
         'V2_20180111_293a_FA27',
         'V2_20170912_T20ABC_FG',
         'V2_20170912_T24ABC_FG',
         'V2_20180111_293a_FL2',
         'V2_20180111_293a_FL26',
 'V3_20180111_HCT116_T16WT',
 'V3_20171215_FG_T17ABC_WT',
 'V3_20171215_FG_T17ABC_FG',
 'V3_20180111_HCT116_T14FG',
 'V3_20171215_FG_T24ABC_FL',
 'V3_20180111_HCT116_T20FL']]
```
